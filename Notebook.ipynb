### **Encode Multi-Label Categorical Columns**

```python
# Use Multi-Hot Coding for multiple diseases or allergies
from sklearn.preprocessing import MultiLabelBinarizer

def is_multiLabel(x):
    if "," in x:
        return [i.strip() for i in x.split()]
    else:
        return [x.strip()]

df['KronikHastalik'] = df['KronikHastalik'].apply(is_multiLabel)
df['Alerji'] = df['Alerji'].apply(is_multiLabel)

mlb_chronic = MultiLabelBinarizer()
chronic_encoded = pd.DataFrame(
    mlb_chronic.fit_transform(df['KronikHastalik']),
    columns=[f'CI_{idx}' for idx in mlb_chronic.classes_]
)

mlb_allergies = MultiLabelBinarizer()

allergy_encoded = pd.DataFrame(
    mlb_allergies.fit_transform(df['Alerji']),
    columns=[f'CI_{idx}' for idx in mlb_allergies.classes_]
)
```

### **Encode Single-Label Categorical Columsn**

```python
# Use OneHotEncoder for single label categories
from sklearn.preprocessing import OneHotEncoder

single_categorical_columns = ['Cinsiyet', 'KanGrubu', 'Bolum', 'Tanilar', 'UygulamaYerleri']

one_hot_encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
single_categorical_encoded = pd.DataFrame(
    one_hot_encoder.fit_transform(df[single_categorical_columns]),
    columns=one_hot_encoder.get_feature_names_out(single_categorical_columns)
)
```
