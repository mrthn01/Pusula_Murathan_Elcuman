### **Plotting Information based on Target Value**

```python
treatment_duration_count = df['TedaviSuresi'].value_counts()

plt.figure(figsize=(8,6))
plt.bar(treatment_duration_count.index.str.replace("Seans", ""), treatment_duration_count, color='blue')
plt.title('Count Plot of Treatment Duration')
plt.xlabel('Seans')
plt.ylabel('Count')
plt.show()
```

```python
sns.histplot(df.TedaviSuresi.str.replace('seans', '', case=False), bins=30, kde=True)
```

```python
# Visualizing Age based on Treatment Session Duration
sns.scatterplot(x='Yas', y='TedaviSuresi', data=df)

plt.xlabel('Age')
plt.ylabel('Treatment Duration')
```

```python
import textwrap

categorical_cols = df.select_dtypes(include=['object']).columns

plt.figure(figsize=(14, len(categorical_cols) * 4))
for i, feature in enumerate(categorical_cols, 1):
    plt.subplot(len(categorical_cols), 2, i)
    sns.countplot(x=df[feature], order=df[feature].value_counts().index, palette="Set2")
    plt.title(f"{feature} | Unique: {df[feature].nunique()}")
    # Fix labels
    labels = [textwrap.fill(label, 10) for label in df[feature].value_counts().index]
    plt.xticks(rotation=45, ha='right')
    plt.gca().set_xticklabels(labels)

plt.tight_layout()
plt.show()
```
